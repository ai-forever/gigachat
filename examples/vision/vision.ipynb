```python
import getpass
import os

if "GIGACHAT_CREDENTIALS" not in os.environ:
    os.environ["GIGACHAT_CREDENTIALS"] = getpass.getpass("Credentials от GigaChat")
```

```python
from gigachat import GigaChat

client = GigaChat(
    base_url="https://gigachat-preview.devices.sberbank.ru/api/v1",
    verify_ssl_certs=False,
    timeout=600,
    model="GigaChat-Pro-preview",
)
```

```python
from IPython.display import Image
Image(filename='cat.jpg')
```

```python
file = client.upload_file(open("cat.jpg", "rb"))
```

```python
result = client.chat(
    {
        "messages": [
            {
                "role": "user",
                "content": "Какой окрас у кота на фото?",
                "attachments": [file.id_],
            }
        ],
        "temperature": 0.1
    }
)
```
