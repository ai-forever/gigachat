# Vision Example

This notebook demonstrates how to use GigaChat API for image analysis using vision capabilities.


Import required libraries for authentication and GigaChat client.

```python
import os
import getpass

from gigachat import GigaChat
```

Set up GigaChat credentials and scope. The input will be hidden for security.

```python
if "GIGACHAT_CREDENTIALS" not in os.environ:
    os.environ["GIGACHAT_CREDENTIALS"] = getpass.getpass("GigaChat Credentials: ")

if "GIGACHAT_SCOPE" not in os.environ:
    os.environ["GIGACHAT_SCOPE"] = getpass.getpass("GigaChat Scope: ")
```

Initialize GigaChat client with a vision-capable model.

```python
client = GigaChat(
    verify_ssl_certs=False,
    timeout=600,
    model="GigaChat-2-Max"
)
```

Display the image that will be analyzed.

```python
from IPython.display import Image
Image(filename='cat.jpg')
```

Upload the image to GigaChat for analysis.

```python
with open("cat.jpg", "rb") as f:
    file = client.upload_file(f)
```

Send a chat request with the uploaded image attachment to analyze its content.

```python
result = client.chat(
    {
        "messages": [
            {
                "role": "user",
                "content": "What color is the cat in the photo?",
                "attachments": [file.id_],
            }
        ],
        "temperature": 0.1,
    }
)

result.choices[0].message.content
```

```python
client.close()
```
