# Simple Chat Example

This notebook demonstrates how to work with chat using GigaChat API with a system message and interactive conversation loop.


Import GigaChat client and required models.

```python
import os
import getpass

from gigachat import GigaChat
from gigachat.models import Chat, Messages, MessagesRole
```

Set up GigaChat credentials and scope. The input will be hidden for security.

```python
if "GIGACHAT_CREDENTIALS" not in os.environ:
    os.environ["GIGACHAT_CREDENTIALS"] = getpass.getpass("GigaChat Credentials: ")

if "GIGACHAT_SCOPE" not in os.environ:
    os.environ["GIGACHAT_SCOPE"] = getpass.getpass("GigaChat Scope: ")
```

## Configure Chat Payload

Set up the initial chat payload with a system message and chat parameters.


```python
payload = Chat(
    messages=[
        Messages(
            role=MessagesRole.SYSTEM,
            content="You are an attentive bot-psychologist who helps the user solve their problems.",
        )
    ],
    temperature=0.7,
    max_tokens=100,
)
```

## Interactive Chat Loop

Start an interactive conversation loop where the user can chat with the bot.


```python
with GigaChat(verify_ssl_certs=False) as giga:
    while True:
        user_input = input("User: ")
        payload.messages.append(Messages(role=MessagesRole.USER, content=user_input))
        response = giga.chat(payload)
        payload.messages.append(response.choices[0].message)
        print("Bot: ", response.choices[0].message.content)
```
