# Context Variables Example

This notebook demonstrates how to forward request headers to GigaChat using context variables. Context variables allow you to set headers and metadata that will be automatically included in API requests.


Import required libraries for context variables and GigaChat client.

```python
import asyncio
import os
import getpass

import gigachat.context
from gigachat import GigaChat
```

Set up GigaChat credentials and scope. The input will be hidden for security.

```python
if "GIGACHAT_CREDENTIALS" not in os.environ:
    os.environ["GIGACHAT_CREDENTIALS"] = getpass.getpass("GigaChat Credentials: ")

if "GIGACHAT_SCOPE" not in os.environ:
    os.environ["GIGACHAT_SCOPE"] = getpass.getpass("GigaChat Scope: ")
```

## Setting Context Variables for Sync Client

Set up context variables that will be used for all requests within the client context. These variables include authorization headers, session IDs, request IDs, and custom headers.


```python
headers = {
    "X-Session-ID": "8324244b-7133-4d30-a328-31d8466e5502",
    "X-Request-ID": "8324244b-7133-4d30-a328-31d8466e5502",
    "X-Service-ID": "my_custom_service",
    "X-Operation-ID": "my_custom_qna",
    "X-Trace-ID": "trace-id-1234567890",
    "X-Agent-ID": "agent-id-1234567890",
}

# Setting context variables for the client
with GigaChat(verify_ssl_certs=False) as giga:
    gigachat.context.session_id_cvar.set(headers.get("X-Session-ID"))
    gigachat.context.request_id_cvar.set(headers.get("X-Request-ID"))
    gigachat.context.service_id_cvar.set(headers.get("X-Service-ID"))
    gigachat.context.operation_id_cvar.set(headers.get("X-Operation-ID"))
    gigachat.context.trace_id_cvar.set(headers.get("X-Trace-ID"))
    gigachat.context.agent_id_cvar.set(headers.get("X-Agent-ID"))
    gigachat.context.custom_headers_cvar.set({"X-Custom-Header": "CustomValue"})

    response = giga.chat("Who are you?")
    print(response)
```

## Setting Context Variables for Async Client

For async operations, you can set context variables per call. This allows different concurrent requests to use different headers. Each async task will have its own context.


```python
async def ask_with_headers(giga, some_custom_header, prompt):
    """Set custom headers for a specific async call."""
    gigachat.context.custom_headers_cvar.set({"X-Custom-Header": some_custom_header})

    response = await giga.achat(prompt)
    print(response)


async def async_main():
    """Run multiple concurrent requests with different custom headers."""
    async with GigaChat(verify_ssl_certs=False) as giga:
        await asyncio.gather(
            ask_with_headers(giga, "CustomValue 1", "Who made you?"),
            ask_with_headers(giga, "CustomValue 2", "What is your name?"),
        )


await async_main()
```
